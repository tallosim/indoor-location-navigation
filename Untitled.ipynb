```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import json
```

```python
data_path = 'data2'
```

```python
ssubm_df = pd.read_csv(os.path.join(data_path, 'output', 'submission_5_1_1.csv'))
ssubm_df[['site', 'path', 'timestamp']] = ssubm_df['site_path_timestamp'].apply(lambda x: pd.Series(x.split('_')))
ssubm_df.drop(['site_path_timestamp'], axis=1, inplace=True)
```

```python
ssubm_df
```

```python
paths = list()
error = 0

for gid, g in ssubm_df.groupby(['path']):
    value_counts = g['floor'].value_counts()
    if value_counts.shape[0] > 1:
        g['floor'] = value_counts.index.tolist()[0]
        error += 1
    paths.append(g)
```

```python
ssubm_df = pd.concat(paths).sort_index()
```

```python
#a path-ek hány százalékában található olyan path, ahol rossz a szint becslés
error/len(paths)
```

```python
ssubm_df['site_path_timestamp'] = ssubm_df['site'] + '_' + ssubm_df['path'] + '_' + ssubm_df['timestamp'] 
ssubm_df.drop(['site', 'path', 'timestamp'], axis=1, inplace=True)

cols = ssubm_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
ssubm_df = ssubm_df[cols]
```

```python
ssubm_df.to_csv(os.path.join(data_path, 'output', 'submission_5_1_1.csv'), index=False)
```

```python
floor_map = {"B2":-2, "B1":-1, "F1":0, "F2": 1, "F3":2, "F4":3, "F5":4, "F6":5, "F7":6,"F8":7, "F9":8}
inverse = {v: k for k, v in floor_map.items()}
```

```python
def visualize_path(path_df):
    site = path_df['site'].iloc[0]
    floor = int(path_df['floor'].iloc[0])
    
    base_path = os.path.join(data_path, 'input', 'metadata', site, inverse[floor])
    
    img = mpimg.imread(os.path.join(base_path,'floor_image.png'))
    
    with open(os.path.join(base_path,'floor_info.json')) as f:
        content = f.read()
        floor_info = json.loads(content)
    height = float(floor_info['map_info']['height'])
    width = float(floor_info['map_info']['width'])
    
    fig = plt.figure(figsize=(15, 12))
    ax = plt.subplot(111)

    plt.imshow(img, extent=[0, width, 0, height])
    
    plt.scatter(path_df['x'], path_df['y'], c='red')
    plt.plot(path_df['x'], path_df['y'], c='red')
```
